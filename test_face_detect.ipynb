```python
# coding: utf-8

# In[1]:


import cv2
import sys
import numpy as np
import os
 
cascPath = sys.argv[1]

faceCascade = cv2.CascadeClassifier("C:\\Users\\a0926_000\\Downloads\\opencv\\sources\\data\\haarcascades_cuda\\haarcascade_frontalface_alt.xml")
faceCascade2 = cv2.CascadeClassifier("C:\\Users\\a0926_000\\Downloads\\opencv\\sources\\data\\haarcascades_cuda\\haarcascade_profileface.xml")

#videoPath = sys.argv[2]
#video_capture = cv2.VideoCapture(videoPath)
video_capture = cv2.VideoCapture(1)
i=0

while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()
    if ret==False:
        break

    cv2.imwrite("D:\\aru\\face"+str(i)+".png",frame)
    i = i+1
    # Display the resulting frame
    cv2.imshow('Video', frame)
 
    if cv2.waitKey(50) & 0xFF == ord('q'):
        break
 # When everything is done, release the capture
video_capture.release()
cv2.destroyAllWindows()

record =  []
img_folder = "D:\\aru"
imglist = os.listdir(img_folder)
count = len(imglist)

for i in imglist:
    flag = 1
    flag1 = 1
    print(img_folder+"\\"+i)
    frame = cv2.imread(img_folder+"\\"+i)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    flipped = cv2.flip(gray,1)
    frame1 = np.zeros(frame.shape)
    
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30),
        flags=cv2.CASCADE_SCALE_IMAGE
    )
    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        record.append({
            "filename": str(i),
            "pos_x":x,
            "pos_y":y,
            "width":w,
            "height":h,
        })
        
        #for j in range(x,x+w):
           # for k in range(y,y+h):
                #if ( j>y & j<y+h & k>x & k<x+w ):
                    #frame1[k][j] = frame[k][j];
        flag = 0
        flag1 = 0 
    if flag:
        faces2 = faceCascade2.detectMultiScale(
            gray,
            scaleFactor=1.1,
            minNeighbors=5,
            minSize=(30, 30),
            flags=cv2.CASCADE_SCALE_IMAGE
        )
        for (x, y, w, h) in faces2:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
            record.append({
                "filename": str(i),
                "pos_x":x,
                "pos_y":y,
                "width":w,
                "height":h,
            })
            #for j in range(x,x+w):
                #for k in range(y,y+h):
                    #if ( j>y & j<y+h & k>x & k<x+w ):
                        #frame1[k][j] = frame[k][j];
            flag1 = 0
        
    if flag1:                   
        faces3 = faceCascade2.detectMultiScale(
            flipped,
            scaleFactor=1.1,
            minNeighbors=5,
            minSize=(30, 30),
            flags=cv2.CASCADE_SCALE_IMAGE
        )
        for (x, y, w, h) in faces3:
            cv2.rectangle(frame, (frame.shape[1]-(w+x), y), (frame.shape[1]-x, y+h), (0, 255, 0), 2)
            record.append({
                    "filename": str(i),
                    "pos_x":frame.shape[1]-(w+x),
                    "pos_y":y,
                    "width":w,
                    "height":h,
            })
l = 0
for i in record:

    info = i
    frame = cv2.imread(img_folder+"\\"+info["filename"])
    frame1 = np.zeros(frame.shape)
    for j in range(info["pos_x"]-20,info["pos_x"]+info["width"]+20):
        for k in range(info["pos_y"]-20,info["pos_y"]+info["height"]+20):
                if j>0 and j<frame1.shape[1] and k>0 and k<frame1.shape[0]:
                    frame1[k][j] = frame[k][j];
    cv2.resize(frame1,(512,512))
    cv2.imwrite(img_folder + "\\mask" + str(l) + ".png",frame1)
    l = l+1
    print(i)
    



```
